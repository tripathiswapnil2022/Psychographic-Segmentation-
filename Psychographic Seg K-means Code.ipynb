```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Psychographic Seg data.xlsx")
```

```python
df.shape
```

```python
data = df.iloc[:, 1:15]
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(init="random", n_clusters=3, n_init=10, max_iter=300, random_state=42)
```

```python
kmeans.fit(data)
```

```python
data.loc[:,"cluster"] = kmeans.labels_
```

```python
df.loc[:,"cluster"] = kmeans.labels_
```
